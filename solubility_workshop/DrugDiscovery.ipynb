#### Keminformatiikan historia

1960-luvulla laskennalliset menetelmät luonnontieteissä, kuten biologiassa alkoivat lisääntymään, ja erityisest Corwin Hanschia ja hänen kollegoitaan voidaan kiittää ensimmäisistä tutkimuksista, jossa huomattiin yhdisteiden ja biologisten vaikutusten perustuvan yhdisteiden rakenteeseen.

Hansch ja muut huomasivat, että yhdisteen pienetkin rakenteelliset muutokset aiheuttivat kasvin auksiinin (kasvuhormoni) toiminnan ja vaikutuksen korostumista. He siis löysivät yhdisteitä, jotka olivat **biologisesti aktiivisia** eli aiheuttivat biologisen vasteen. 

Tämä aineen rakenteen suhde biologiaan oli matemaattisesti todettavissa korrelaationa, eli muuttujien keskinäisenä riippuvuutena. Lisäksi aineen pitoisuus vaikutti myös biologiseen vasteeseen. **Syntyi teoria rakenteen ja aktiivisuuden yhteydestä**, joka voitiin ilmaista laskennallisesti. Tuloksista syntyi malli, jolla voidaan ennustaa aineen ominaisuuksia. 

Hansch nimesi sen kvantitatiiviseksi rakenne-aktiivisuussuhteeksi (engl. _Quantitative Structure-activity-relationship_, QSAR). 



**Kuva 1.** QSAR-malleja käytetään ennustamaan rakenteen perusteella ominaisuuksia esimerkiksi lääketutkimuksia varten. 

Tutkijat innostuivat Hanschin tuloksista tutkimaan myös muita tapauksia, kuin auksiinin aktiivisuuteen vaikuttavia yhdisteitä, ja huomatiinkin nopeasti laajan kirjon lääkkeitä vaikuttavan rakenteesta johdetuilla laskennallisilla menetelmillä biologista aktiivisuutta.

Kemiallista dataa syntyi kuitenkin paljon, ja vain tietokoneilla aineistoja voitiin käsitellä kustannustehokkaasti. Täten 1960-1970-luvulla kemistit ja tietojenkäsittelytieteilijät alkoivat  tutkimaan miten kemiallista dataa voidaan parhaiten esittää ja varastoida tietoteknisesti.

### Yhdisteiden esittäminen tietokoneille: SMILES

Oletko ikinä miettinyt, miten tietokone ymmärtää piirtämiäsi molekyylejä? 

Yhdisteitä, eli atomeja ja niiden välisiä sidoksia, on pyritty kuvaamaan matemaattisella kielellä tietokoneelle erilaisilla menetelmillä. Tämä osoittautui erityisen haastavaksi matemaattiseksi mallinnustyöksi, jossa erilaiset työläät ja puutteelliset tavat viestiä tietokoneelle yhdisteistä vasta useiden työläiden ja usein puutteellisten tietoteknisten yritysten jälkeen, 1980-luvulla kolmen veljeksen yritys Daylight informtics onnistui kehittämään tiedostotyypin nimeltä: SMILES.

SMILES (Simplified Molecular Input Line Entry System) on ylivertaisen yksinkertainen tapa viestiä molekyylirakenteesta yksinkertaisella, mutta kattavalla tavalla kirjoittaa, lukea ja tallentaa tietoteknisiä kuvauksia kemiallisista yhdisteistä. SMILES koostuu alkuaineiden ja numeroiden sarjasta, jota luetaan tietyllä tavalla muodostaen, graafi, eli pisteiden ja janojen verkko. Tämä verkko nojaa matemaattiseen verkkoteoriaan, ja se rakennetaan erilaisia algoritmeilla ja sääntötaulukolla.

Kanoninen SMILES:

- Kanoninen SMILES on ainutlaatuinen merkkijono molekyylirakenteesta.
- Hyödyllinen erityisesti kun etsitään tarkkaa rakennetta mistä tahansa datasetistä, sillä variantteja ei ole.

Isomeerinen SMILES:

- Nimensä mukaisesti ottaa huomioon yhdisteessä isomerian.
- Jos yhdiste on enantiomeeri, vain isomeerinen SMILES voi erottaa yhdisteet toisistaan, vaikka niillä on sama molekyylikaava.

#### PUG-REST menetelmä

Ohjelmistoympäristöön voi tuoda helposti yhdisteitä teettämällä "kyselyitä" PubChem-palvelimelta. Tämä on erityisen hyödyllistä, kun halutaan nopeasti tuoda tietoa lääkeyhdisteistä ilman hakukoneita luettelemalla haluttuja molekyylejä osaksi koodia.

```python
# Tee lista tuntemistasi lääkemolekyyleistä tähän englannin kielellä ja erota jokainen nimike pilkulla ja lainausmerkeillä seuraavanlaisesti.

# Esimerkiksi, kirjoita englannin kielellä.
compound_names = ["Aspirin","Paracetamol", "Venlafaxin", "Thalidomide", "Benzamide", "Cytosine"] 
```

```python
import requests
 # Älä muokkaa alla olevaa koodia. 
def retrieve_names(compound_names):
    pugrest = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
    pugoper = "property/CanonicalSMILES"
    pugout = "txt"
    results = {}
    for compound_name in compound_names:
        pugin = "compound/name/" + compound_name
        url = f"{pugrest}/{pugin}/{pugoper}/{pugout}"
        response = requests.get(url)
        if response.status_code == 200:
            results[compound_name] = response.text.strip()  # Strip to remove leading/trailing whitespaces
        else:
            results[compound_name] = f"Error: {response.status_code}"
    return results

# Muokkaa vain alla olevaa koodiriviä.
compound_names = ["Aspirin","Paracetamol", "Venlafaxin", "Cocaine", "Benzamide", "Cytosine"] 
#Muokkaa vain yllä olevaa koodiriviä.
canonical_smiles = retrieve_names(compound_names)
for compound_name, smiles in canonical_smiles.items():
    print(f"Canonical SMILES for {compound_name}: {smiles}")
```

```python
# Voit myös muuttaa SMILES merkkijonon isomeeriseksi seuraavalla komentosarjalla. 
from rdkit import Chem
from rdkit.Chem import MolFromSmiles, MolToSmiles

def canonical_to_isomeric(smiles):
    mol = MolFromSmiles(smiles)
    if mol is not None:
        Chem.SanitizeMol(mol)
        isomeric_smiles = MolToSmiles(mol, isomericSmiles=True)
        return isomeric_smiles
    else:
        return None

# Muokkaa vain alla olevaa riviä
canonical_smiles = "CN1C2CCC1C(C(C2)OC(=O)C3=CC=CC=C3)C(=O)OC"
isomeric_smiles = canonical_to_isomeric(canonical_smiles)
print("Canonical SMILES:", canonical_smiles)
print("Isomeric SMILES:", isomeric_smiles)
```

### Lääkeyhdisteiden ominaisuudet kehossa

Lääkeyhdisteiden vaikutuksia ja vaikutusten perustaa on ollut kuitenkin hankalaa tieteellisesti tutkia. 1800-luvulla oli tavanomaista nojata hyvin kokeelliseen lääkeyhdisteiden etsimiseen vapaaehtoisilla sairaiilla potilailla. Lääkärien toiminta perustui pitkälle rohtojen ja yrttien sekä muitten hoitojen perimätietoon, omaperäisiin tulkintoihin ja kokemukseen sekä väliajoin heidän työtään ohjasi esimiehen, usein tutkivan ylilääkärin auktoriteett. Vasta 1800-luvun puolivälissä kemialliset erotus- ja analyysimenetelmät sekä käsite funktionaalisista ryhmistä eli rakenteiden vaikutuksesta kemiallisiin ominaisuuksiin edisti ymmärrystä tiettyjen yhdisteiden vaikutuksista.

#### ADME-ominaisuudet 

Kemialtaan lääkeaineiden biokemialliset vaikutukset jaetaan neljään ryhmään: imeytyminen, jakautuminen, metabolia ja erittyminen. Nämä muodostavat englannin kiellä lyhenteen ADME (_Adsorbtion, distribution, metabolisim and elimination_). 

Yksi merkityksellisimmistä ominaisuuksista on lääkkeen **imeytyminen**, eli miten nautittu aine imeytyy verenkiertoon. Erilaisia keinoja on useita, mutta keskeisimmät lienevät:

1. Oraalinen: nautitaan suunkautta, mutta kulkeutuminen kehoon voi tapahtua läpäisemällä joko suolistoepiteelin  Helppo ja tyypillisin tapa saada kehoon lääkeainetta. Suositaan erityisesti, kun lääkeainetta halutaan imeytyvän verenkiertoon hitaasti tai vain osa alkuperäisestä massasta. Lisäksi maksan ja suoliston kyky muokata metabolisesti entsyymeillään lääkeyhdistettä ennen suureen verenkiertoon imeytymistä on merkityksellistä esim. venlafaksiinissa.
2. Posken limakalvon tai kielenalaisen kudosepiteelin läpäiseminen on nopeampi keino, mutta voi olla myös haitallinen nopean lääkeaineen pitoisuuden nousun myötä veressä.
3. Peräsuoli: lääke saadaan verenkiertoon verisuonitiheän sulkijalihaksen limakalvon kautta.

Voimme päätellä että erityisesti limakalvot ovat merkityksellisiä imeytymisen kannalta, mutta ohuen kudosesteen lisäksi lääkeaineella täytyy olla tietty ominaisuus: **hyvä vesiliukoisuus**

Muita lääkeyhdisteiden ominaisuuksia on **jakautuminen** kehossa, joka on mitta lääkepitoisuuden jakautumisesta elimissä ja kudoksissa. Tähän vaikutta hyvin moni asia, mutta keskeisimpiä tekijöitä ovat tiettyjen biologisten esteiden kuten aivo-veriesteen tai proteiinien ja rasvojen pitoisuuksien vaikutus.

**Metabolia** on merkittävä askel imeytymisen ja jakautumisen jälkeen, kun keho kohtelee lääkeainetta kuona-aineena poistaen sen joko ulosteen tai virtsan mukana. Maksa ja sen entsyymit vastaavat tästä prosessista.

**Erittyminen** on prosessi, jossa paksusuoli ja munuaiset poistavat lääkeainetta verenkierrosta. Tämä voi olla erityisen vaikeaa rasvaliukoisten yhdisteiden tai raskasmetalleja sisältävien aineiden kanssa.

### Mikä on hyvä lääke?

Kysymys on ollut hyvin kokeellinen lääkekehityksessä kautta aikojen, ja kokeellisuudella pärjätään vieläkin pisimmälle. Ihmiset ovat esimerkiksi pajujen kaarnaa pureskelemalla tai Sechuanin pippureita maistelemalla saaneet kivunlievitystä erilaisiin vaivoihin. Antiobiootit löytyivät lähes sattumalta, mutta kokeilemalla niitä on löydetty uusia. Ihmiskunta tuntee jo tuhansia lääkkeitä, ja jälleen kerran informaatiotietieden hengessä keminformaatikkoja ja lääkäreitä on kiinnostanut: mikä niitä yhdistää?

Tähän kysymykseen Lipinksin johtama tutkimusryhmä etsi vastausta aloittaen suuren metatutkimuksen erilaisista patentoiduista lääkkeistä:

#### Lipinskin viiden sääntö

Christopher Lipinski tutki tiiminsä kanssa n. 2000 lääkkeen dataa ja osoitti, että niillä on yleisesti päteviä hyviä piirteitä: 

- Molekyylimassa on alle 500 Da, jossa 1 Da vastaa noin 1 atomimassayksikkö 
- Jakautumiskerroin logP on alle 5
- Vetydonorien lukumäärä on alle 5, jossa vetydonori on atomi/rakenne, joka vetysidoksen luovuttuaan toimii negatiivisella osittaisvarauksella (Esim. happi, O vesimolekyylien välisissä vetysidoksissa)
- Vetyakseptorien lukumäärä on alle 10, eli vetydonorien vastakohta siis pos. osittaisvarauksellinen atomi

Saatiin selville että useimmat lääkkeet poikkesivat vain vähän näistä arvoista korkealla luotettavuudella. Toisaalta Lipinski ym. eivät ottaneet huomioon luonnonyhdisteitä, ja moni luonnonyhdiste on todettu erittäin hyödylliseksi ja Lipisnkin sääntöjä rikkovaksi.


#### Lipinski ei ole aina oikeassa

Nykyään tieteilijät myös kritisoivat Lipinskin viiden sääntöä siitä, että se painottaa jokaista sääntöä yhtä suuresti. Tätä tukevcaa poikkeavaa ajattelua on olemassa myössen takia, että jotkin yhdisteet toimivat loistavina lääkkeinä, vaikka ne rikkovat Lipinskin viiden sääntöä.

**Daptomysiini**: Daptomysiini on lipopeptidiantibiootti, jota käytetään vakavien bakteeri-infektioiden hoitoon. Sen molekyylipaino on suhteellisen suuri (yli 1000 Da) ja korkea LogP-arvo sen rasvaliukoisen luonteen vuoksi. Näistä ominaisuuksista huolimatta daptomysiini pysyy tehokkaana, koska se vaikuttaa pääasiassa infektion kohteena olevassa kohdassa, missä se pystyy tunkeutumaan bakteerisolukalvoihin ja vaikuttamaan antibakteerisesti.

**Vancomysiini**: Vancomysiini on toinen antibiootti, joka ylittää Lipinskin molekyylipainon suosituksen. Se on glykoproteiiniantibiootti, jota käytetään Gram-positiivisten bakteerien aiheuttamien infektioiden hoitoon. Vancomysiini toimii estämällä soluseinän synteesiä bakteereissa. Sen suuri koko ja hydrofiilinen luonne eivät estä sen aktiivisuutta kohdebakteereita vastaan, kun sitä annetaan asianmukaisesti.

**Siklosporiini**: Siklosporiini on immunosuppressiivinen lääke, jota käytetään hylkimisen estoon elinsiirroissa sekä autoimmuunisairauksien, kuten nivelreuman ja psoriasiksen, hoitoon. Siinä on suuri molekyylipaino ja LogP-arvo, mutta se on muotoiltu tavalla, joka mahdollistaa tehokkaan toimituksen kohdekudoksiin, kuten immuunivasteeseen osallistuviin lymfosyytteihin.

**Takrolimuusi**: Samoin kuin siklosporiini, takrolimuusi on immunosuppressiivinen lääke, jota käytetään elinsiirroissa ja autoimmuunisairauksissa. Siinä on ominaisuuksia, jotka poikkeavat Lipinskin säännöistä, mutta se on muotoiltu saavuttamaan terapeuttiset pitoisuudet immuunisoluissa samalla kun se minimoi systeemiset sivuvaikutukset.

#### Lääkkeiden ominaisuuksiin vaikuttavat tekijät

Lääkeyhdisteet ovat yleensä orgaanisen kemian yhdisteisiin. Orgaanisten yhdisteiden keskeinen piirre on hiiliatomien väliset kovalenttiset sidokset, jotka voivat olla sidostyypiltää yksinkertaisia sidoksia, kaksoissidoksia tai kolmoissidoksia.

- Jakautumiskerroin logP (kerroin joka kuvaa liukenemistaipumusta joko rasvaliukoiseen tai vesiliukoiseen faasiin)
- Moolimassa vaikutti kääntäenverrannollisesti vesiliukoisuuteen, eli raskaat orgaaniset yhdisteet eivät liukene hyvin veteen.
- Pyörivät sidokset ovat ei rengasrakenteissa sijaitsevia sp3 hybridisoituneita sidoksia, jotka voivat pyöriä vapaasti akselinsa ympäri
- Aromaattisuusaste eli kuinka monta atomia ovat aromaattia suhteessa kaikkiin atomeihin.

### Regressioanalyysi

Lineeariset sovitukset ovat tuttuja varmasti matematiikan tunneilta. Tämä matemaattinen mallintaminen on yksi keskeisimpiä menetelmiä keminformatiikassa koska sillä voidaan tarkastella erilaisten muuttjien suhdetta toisiinsa (korrelaatio) aloittaen alkaen kahdesta tai usemmasta tekijästä. Alla on koodia, jolla voidaan sovittaa regressiomalli alkaanien ja sykloalkaaninien datasettiin. Mitä huomaat kun koodi tulostaa graafin? Onko tämä tarkka malli?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Lataa dataa .csv tiedostosta taulukoksi
df = pd.read_csv('BP.csv', encoding='latin-1')

# Sovita lineaarinen regressiomalli 
model = LinearRegression()
X = df[['Moolimassa']]  # Feature (independent variable)
y = df['Kiehumispiste_K']  # Target variable (dependent variable)
model.fit(X, y)

# Piirrä  pistekuvio datasta 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Moolimassa', y='Kiehumispiste_K', data=df, color='blue', label='Data Points')

# Piirrä regressiosuora
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

# Aseta otsikot ja mitat
plt.xlabel('Moolimassa')
plt.ylabel('Kiehumispiste (K)')
plt.title('Linear Regression Model')

# Show legend
plt.legend()

# Näytä sovitus
plt.show()
```
